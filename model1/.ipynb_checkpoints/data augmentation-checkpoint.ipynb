```python
import random
import json


# File path
file_path = 'GoodBadVariableNames_JS.json'  # Update this with the actual path to your data.json file

# Initialize lists
bad_snippets = []
good_snippets = []
# Read JSON data from the file
with open(file_path, 'r') as file:
    data = json.load(file)
    for item in data:
        bad_snippets.append(item['bad'])
        good_snippets.append(item['good'])
        
        
def generate_dataset(bad_snippets, good_snippets, num_examples):
    dataset = []
    half_examples = num_examples // 2

    for _ in range(half_examples):
        # Generate 'bad' examples
        num_bad = random.randint(1, 3)  # At least one bad snippet
        num_good = random.randint(0, 2)  # Random number of good snippets
        snippets = random.choices(bad_snippets, k=num_bad) + random.choices(good_snippets, k=num_good)
        random.shuffle(snippets)
        dataset.append({"code": "\n".join(snippets), "label": "bad"})

    for _ in range(half_examples, num_examples):
        # Generate 'good' examples
        num_snippets = random.randint(1, 4)
        snippets = random.choices(good_snippets, k=num_snippets)
        dataset.append({"code": "\n".join(snippets), "label": "good"})

    return dataset



# Generate the dataset
dataset = generate_dataset(bad_snippets, good_snippets, num_examples=30000)

# Print the first few examples to check
for example in dataset[:5]:
    print(example)
# Save the dataset to a JSON file
output_file_path = 'augmented_dataset2.json'
with open(output_file_path, 'w') as file:
    json.dump(dataset, file, indent=4)
```

```python
import random
import json

# File path
file_path = 'clean.json'  # Update this with the actual path to your data.json file

# Initialize lists
bad_snippets = []
good_snippets = []

# Read JSON data from the file
with open(file_path, 'r') as file:
    data = json.load(file)
    for item in data:
        bad_snippets.append(item['bad'])
        good_snippets.append(item['good'])

def generate_dataset(bad_snippets, good_snippets, num_examples):
    dataset = []

    for _ in range(num_examples):
        # Randomly select one bad and one good snippet
        bad_snippet = random.choice(bad_snippets)
        good_snippet = random.choice(good_snippets)

        dataset.append({"bad": bad_snippet, "good": good_snippet})

    return dataset

# Generate the dataset
dataset = generate_dataset(bad_snippets, good_snippets, num_examples=3000)

# Save the dataset to a JSON file
output_file_path = 'augmented_dataset2.json'
with open(output_file_path, 'w') as file:
    json.dump(dataset, file, indent=2)
```

```python
import json
import re

def remove_comments(code):
    # Regular expression to remove single-line comments (// ...) and multi-line comments (/* ... */)
    pattern = r"(\/\/[^\n]*|\/\*[\s\S]*?\*\/)"
    return re.sub(pattern, '', code)

def process_json_file(input_file, output_file):
    try:
        # Read the input JSON file
        with open(input_file, 'r') as file:
            data = json.load(file)

        # Process the data by removing comments
        for entry in data:
            entry['bad'] = remove_comments(entry['bad'])
            entry['good'] = remove_comments(entry['good'])

        # Save the processed data to the output JSON file
        with open(output_file, 'w') as file:
            json.dump(data, file, indent=4)

        print(f"Comments removed and saved to {output_file}")

    except Exception as e:
        print(f"An error occurred: {str(e)}")

# Specify the input and output file paths
input_file_path = 'GoodBadVariableNames_JS.json'  # Replace with the path to your input JSON file
output_file_path = 'clean.json'  # Replace with the desired output file path

# Process the JSON file and remove comments
process_json_file(input_file_path, output_file_path)
```
